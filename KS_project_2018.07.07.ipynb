```python
import pandas as pd 
 
Location = "music.csv" #select music
df = pd.read_csv(Location)
df.head()
```

```python
def score_to_numeric(x):
    if x=='successful':
        return 1
```

```python
df['WIN'] = df['state'].apply(score_to_numeric)
df.tail(10)
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='WIN ~ goal + backers - 1', data=df).fit()
result.summary()
```
