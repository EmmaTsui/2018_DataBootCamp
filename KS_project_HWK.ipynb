```python
df.loc[df['state'] == "successful", 'backers'].mode() 
# when backers = 17 kickstarters become far more likely to be funded
```

```python
df.loc[df['state'] == "successful", 'goal'].sum() #237854743.16000006
df.loc[df['state'] == "successful", 'goal'].mean() #10162.995349512907
df.loc[df['state'] == "successful", 'goal'].max() #1400000.0
df.loc[df['state'] == "successful", 'goal'].min() # 1.0
df.loc[df['state'] == "successful", 'goal'].std() #29500.554753841814
df.loc[df['state'] == "successful", 'goal'].mode() # 5000.0
df.loc[df['state'] == "successful", 'goal'].quantile(.5) #3894.0
df.loc[df['state'] == "successful", 'goal'].quantile(.25) #1250.0
df.loc[df['state'] == "successful", 'goal'].quantile(.75) #10000.0
# when goal = 10000.0 which become far less likely to be reached.
```
